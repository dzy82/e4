```python
import random
import time

# Card definitions
card_deck = [
    {"name": "Diablo", "health": 100, "attack": 90, "defense": 60},
    {"name": "Medusa", "health": 100, "attack": 70, "defense": 70},
    {"name": "Jester", "health": 120, "attack": 60, "defense": 90},
    {"name": "Troll", "health": 150, "attack": 40, "defense": 94},
    {"name": "Specter", "health": 100, "attack": 70, "defense": 70},
    {"name": "Mist", "health": 100, "attack": 75, "defense": 65},
    {"name": "Savage", "health": 100, "attack": 90, "defense": 50},
    {"name": "Marauder", "health": 100, "attack": 85, "defense": 50},
    {"name": "Wimp", "health": 110, "attack": 40, "defense": 85},
    {"name": "Sorcerer", "health": 100, "attack": 70, "defense": 55}
]

deck_size = 5

def draw_random_deck(deck, size):
    return random.sample(deck, size)

def print_card(card):
    print(f"Name: {card['name']}\nHealth: {card['health']}\nAttack: {card['attack']}\nDefense: {card['defense']}\n")

def select_card(deck):
    while True:
        try:
            print("Choose a card by number:")
            for i, card in enumerate(deck):
                print(f"{i + 1}. {card['name']}")
            choice = int(input()) - 1
            if 0 <= choice < len(deck):
                return deck[choice]  # Don't remove it from the deck, just return the card
            else:
                print("Invalid choice. Please select a valid number.")
        except ValueError:
            print("Invalid input. Please enter a number.")

def attack_card(attacker, defender):
    print(f"{attacker['name']} attacks {defender['name']}.")
    
    # 先计算对防御的伤害
    damage_to_defense = min(defender['defense'], attacker['attack'])
    defender['defense'] -= damage_to_defense
    print(f"{defender['name']}'s defense is reduced to {defender['defense']}.")
    
    # 剩余的攻击力转化为对生命值的伤害
    remaining_attack = attacker['attack'] - damage_to_defense
    if remaining_attack > 0:
        defender['health'] -= remaining_attack
        print(f"{defender['name']}'s health is reduced to {defender['health']}.")

def game_turn(player_deck, opponent_deck):
    if not player_deck:
        print("Player has no cards left. Opponent wins!")
        return False
    if not opponent_deck:
        print("Opponent has no cards left. Player wins!")
        return False

    print("\nPlayer's Deck:")
    for i, card in enumerate(player_deck):
        print(f"{i + 1}. {card['name']}")

    opponent_card = random.choice(opponent_deck)
    print("\nOpponent's Card:")
    print_card(opponent_card)

    player_card = select_card(player_deck)
    print("\nPlayer's Chosen Card:")
    print_card(player_card)

    if player_card['attack'] > opponent_card['health']:
        print(f"{player_card['name']} defeats {opponent_card['name']}!")
        opponent_deck.remove(opponent_card)
    else:
        attack_card(player_card, opponent_card)
        if opponent_card['health'] <= 0:
            print(f"{opponent_card['name']} is defeated!")
            opponent_deck.remove(opponent_card)

    # Check if the player card should be removed
    if player_card['health'] <= 0:
        print(f"{player_card['name']} is defeated!")
        player_deck.remove(player_card)

    time.sleep(3)  # Pause for a while

    if not opponent_deck:
        print("Opponent has no cards left. Player wins!")
        return False

    print("\nOpponent's Turn:")
    opponent_card = random.choice(opponent_deck)
    print("Opponent's Card:")
    print_card(opponent_card)

    player_card = random.choice(player_deck)
    print("Player's Card:")
    print_card(player_card)

    if player_card['attack'] > opponent_card['health']:
        print(f"Opponent's {player_card['name']} defeats {opponent_card['name']}!")
        opponent_deck.remove(opponent_card)
    else:
        attack_card(opponent_card, player_card)
        if player_card['health'] <= 0:
            print(f"{player_card['name']} is defeated!")
            player_deck.remove(player_card)

    print("A new round will begin...")
    time.sleep(3)  # Pause for a while
    return True

# Initialize decks
player_deck = draw_random_deck(card_deck, deck_size)
opponent_deck = draw_random_deck(card_deck, deck_size)

# Game loop
game_active = True
while game_active:
    game_active = game_turn(player_deck, opponent_deck)

if not player_deck:
    print("Player has no cards left. Opponent wins!")
else:
    print("Player wins!")



```
